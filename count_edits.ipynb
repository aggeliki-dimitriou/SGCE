```python
import os
import os.path as osp
import cv2
import matplotlib.pyplot as plt


class CUB:

    def __init__(self, dataset_path):
        self.dataset_path = dataset_path
        self.load_data()


    def load_data(self):

        # =========================
        # IMAGES AND CLASS LABELS:
        # =========================
        with open(osp.join(self.dataset_path, "images.txt"), "r") as handle:
            self.image_id_to_image_name = {int (line.strip().split(" ")[0]): line.strip().split(" ")[1] for line  in handle.readlines()}

        with open(osp.join(self.dataset_path, "train_test_split.txt"), "r") as handle:
            self.image_id_to_train_label = {int(line.strip().split(" ")[0]): int(line.strip().split(" ")[1]) for line in handle.readlines()}

        with open(osp.join(self.dataset_path, "classes.txt"), "r") as handle:
            self.class_id_to_class_name = {int(line.strip().split(" ")[0]): line.strip().split(" ")[1] for line in handle.readlines()}

        with open(osp.join(self.dataset_path, "image_class_labels.txt"), "r") as handle:
            self.class_id_to_class_name = {int(line.strip().split(" ")[0]): int(line.strip().split(" ")[1]) for line in handle.readlines()}

        # =========================
        # PART LOCATIONS:
        # =========================

        with open(osp.join(self.dataset_path, "parts/parts.txt"), "r") as handle:
            self.part_id_to_part_name = {int(line.strip().split(" ")[0]): " ".join(line.strip().split(" ")[1:]) for line in handle.readlines()}

        self.image_id_to_parts = {}
        with open(osp.join(self.dataset_path, "parts/part_locs.txt"), "r") as handle:
            for line in handle.readlines():
                image_id, part_id, _, _, visible = line.strip().split(" ")
                if int(visible) == 0:
                    continue

                part_name = self.part_id_to_part_name[int(part_id)]
                if int (image_id) in self.image_id_to_parts:
                    self.image_id_to_parts[int(image_id)].append(part_name)
                else:
                    self.image_id_to_parts[int(image_id)] = [part_name]

        # =========================
        # ATTRIBUTE LABELS:
        # =========================
        with open(osp.join(self.dataset_path, "attributes/attributes.txt"), "r") as handle:
            self.attribute_id_to_attribute_name = {int(line.strip().split(" ")[0]): line.strip().split(" ")[1] for line in handle.readlines()}

        with open(osp.join(self.dataset_path, "attributes/certainties.txt"), "r") as handle:
            self.certainty_id_to_certainty_name = {int(line.strip().split(" ")[0]): line.strip()[2:] for line in handle.readlines()}

        self.image_id_to_annotations = {}
        with open(osp.join(self.dataset_path, "attributes/image_attribute_labels.txt"), "r") as handle:
            for line in handle.readlines():
                try:
                    image_id, attribute_id, is_present, certainty_id, time =  line.strip().split(" ")
                except Exception as e:
                    continue

                if int(is_present) == 0:
                    continue
                certainty_name = self.certainty_id_to_certainty_name[int(certainty_id)]
                attribute_name = self.attribute_id_to_attribute_name[int(attribute_id)]

                if certainty_name not in ["probably", "definitely"]:
                    continue

                # if float (time) < 10:
                #     continue

                if int (image_id) in self.image_id_to_annotations:
                    self.image_id_to_annotations[int (image_id)].append(attribute_name)
                else:
                    self.image_id_to_annotations[int (image_id)] = [attribute_name]

    def print_annotations(self, image_id):
        print (self.image_id_to_annotations[image_id])

    def print_parts(self, image_id):
        print (self.image_id_to_parts[image_id])

    def print_sample(self, image_id):
        image_name = self.image_id_to_image_name[image_id]
        image_path = osp.join(self.dataset_path, "images", image_name)
        image = cv2.imread(image_path)
        self.print_parts(image_id)
        self.print_annotations(image_id)
        plt.imshow(image)

    def get_info(self, image_id):
      return self.image_id_to_parts[image_id], self.image_id_to_annotations[image_id]

    def get_image_ids(self):
      return list(self.image_id_to_image_name.keys())

    def get_id(self, name):
      for k, v in self.image_id_to_image_name.items():
        if v == name:
          return k


!mv /content/attributes.txt /content/CUB_200_2011/attributes/attributes.txt
dataset = CUB('CUB_200_2011')
```

```python
import matplotlib.pyplot as plt
from networkx.drawing.nx_pydot import graphviz_layout

def draw_scenegraph(graph):
  node_labels = nx.get_node_attributes(graph, "label")
  edge_labels = nx.get_edge_attributes(graph, "label")


  pos = graphviz_layout(graph, prog='twopi')
  plt.figure(figsize=(10, 10))
  nx.draw(
      graph, pos, edge_color='black',
      node_size=2000, node_color='pink', alpha=0.9,
      labels=node_labels
  )
  nx.draw_networkx_edge_labels(
      graph, pos,
      edge_labels=edge_labels,
      font_color='red'
  )

  plt.axis('off')
  plt.show()
```

```python
def count_edits(key, node_mapping, edge_mapping):
  pair1 = key
  pair2 = (key[1], key[0])
  node_edits1 = 0
  node_edits2 = 0
  edge_edits1 = 0
  edge_edits2 = 0

  cub_idx_ = [i[0] for i in cub_idx]
  synn1 = syn_n[cub_idx_.index(key[0])]
  syne1 = syn_e[cub_idx_.index(key[0])]

  synn2 = syn_n[cub_idx_.index(key[1])]
  syne2 = syn_e[cub_idx_.index(key[1])]

  nodes1 = list(cub_graphs[cub_idx_.index(key[0])].nodes())
  nodes2 = list(cub_graphs[cub_idx_.index(key[1])].nodes())
  edges1 = list(cub_graphs[cub_idx_.index(key[0])].edges())
  edges2 = list(cub_graphs[cub_idx_.index(key[1])].edges())

  node_edits1_ = []
  node_edits2_ = []
  edge_edits1_ = []
  edge_edits2_ = []

  for i in range(len(node_mapping[0])):
    # print(node_mapping[0])
    if node_mapping[0][i] != i:
      if node_mapping[0][i] != None:
        try:
          if synn1[nodes1[i]] != synn2[nodes2[node_mapping[0][i]]]:
            node_edits1+=1
            node_edits1_.append((synn1[nodes1[i]][0], synn2[nodes2[node_mapping[0][i]]][0]))
        except:
            node_edits1+=1
            node_edits1_.append((None, synn2[nodes2[node_mapping[0][i]]][0]))
      elif node_mapping[0][i] == None:
        node_edits1+=1
        node_edits1_.append((synn1[nodes1[i]][0], None))

  for i in range(len(node_mapping[1])):
    if node_mapping[1][i] != i:
       if node_mapping[1][i] != None:
        try:
          if synn2[nodes2[i]] != synn1[nodes1[node_mapping[1][i]]]:
            node_edits2+=1
            node_edits2_.append((synn2[nodes2[i]], synn1[nodes1[node_mapping[1][i]]]))
        except:
            node_edits2+=1
            # print(node_mapping[1], len(node_mapping[1]), len(nodes1), len(nodes2))
            node_edits2_.append((None, synn1[nodes1[node_mapping[1][i]]]))
       elif node_mapping[1][i] == None:
        node_edits2+=1
        node_edits2_.append((synn2[nodes2[i]], None))

  for i in range(len(edge_mapping[0])):
    if edge_mapping[0][i] != i:
      if edge_mapping[0][i] != None:
        try:
          if syne1[edges1[i]] != syne2[edges2[edge_mapping[0][i]]]:
            edge_edits1+=1
            edge_edits1_.append((syne1[edges1[i]], syne2[edges2[edge_mapping[0][i]]]))
        except:
          edge_edits1+=1
          edge_edits1_.append((None, syne2[edges2[edge_mapping[0][i]]]))
      elif edge_mapping[0][i] == None:
        edge_edits1+=1
        edge_edits1_.append((syne1[edges1[i]], None))

  for i in range(len(edge_mapping[1])):
    if edge_mapping[1][i] != i:
      if edge_mapping[1][i] != None:
        try:
          if syne2[edges2[i]] != syne1[edges1[edge_mapping[1][i]]]:
            edge_edits2+=1
            edge_edits2_.append((syne2[edges2[i]], syne1[edges1[edge_mapping[1][i]]]))
        except:
          edge_edits2+=1
          edge_edits2_.append((None, syne1[edges1[edge_mapping[1][i]]]))
      elif edge_mapping[1][i] == None:
        edge_edits2+=1
        edge_edits2_.append((syne2[edges2[i]], None))

  return node_edits1, node_edits2, edge_edits1, edge_edits2 , {'node1': node_edits1_, 'node2': node_edits2_, 'edge1': edge_edits1_, 'edge2': edge_edits2_}
```

```python
def find_edits(rank):
  edits = {}
  edits_path = {}
  for i in range(160):
    q_idx = cub_idx[262:422][i][0]
    pair = (q_idx, rank[i][0])

    if pair in geds_cub.keys():
      _, nodem, edgem = geds_cub[pair]
      n, _, e, _ , d= count_edits(pair, nodem, edgem)
    elif (pair[1], pair[0]) in geds_cub.keys():
      # print('yes')
      pair_ = (pair[1], pair[0])
      _, nodem, edgem = geds_cub[pair_]
      _, n, _, e, d = count_edits(pair_, nodem, edgem)
    else:
      continue

    edits[pair] = (n,e)
    edits_path[pair] = d

  return edits, edits_path
```

```python
def count_edits(key, node_mapping, edge_mapping, cub_idx_, cub_graphs):
  pair1 = key
  pair2 = (key[1], key[0])
  node_edits1 = 0
  node_edits2 = 0
  edge_edits1 = 0
  edge_edits2 = 0

  G1 = cub_graphs[cub_idx_.index(key[0])]
  G2 = cub_graphs[cub_idx_.index(key[1])]

  nodes1 = list(G1.nodes())
  nodes2 = list(G2.nodes())
  edges1 = list(G1.edges())
  edges2 = list(G2.edges())

  synn1 = {node: [G1.nodes()[node]['label']] for node in nodes1}
  syne1 = {edge: G1.edges()[edge]['label'] for edge in edges1}

  synn2 = {node: [G2.nodes()[node]['label']] for node in nodes2}
  syne2 = {edge: G2.edges()[edge]['label'] for edge in edges2}

  node_edits1_ = []
  node_edits2_ = []
  edge_edits1_ = []
  edge_edits2_ = []

  for i in range(len(node_mapping[0])):
    # print(node_mapping[0])
    if node_mapping[0][i] != i:
      if node_mapping[0][i] != None:
        try:
          if synn1[nodes1[i]] != synn2[nodes2[node_mapping[0][i]]]:
            node_edits1+=1
            node_edits1_.append((synn1[nodes1[i]][0], synn2[nodes2[node_mapping[0][i]]][0], [((synn1[e[0]][0], synn1[e[1]][0]), syne1[e]) for e in edges1 if nodes1[i] in e]))
        except:
            node_edits1+=1
            node_edits1_.append((None, synn2[nodes2[node_mapping[0][i]]][0], [((synn1[e[0]][0], synn1[e[1]][0]), syne1[e]) for e in edges1 if nodes1[i] in e]))
      elif node_mapping[0][i] == None:
        node_edits1+=1
        node_edits1_.append((synn1[nodes1[i]][0], None, [((synn1[e[0]][0], synn1[e[1]][0]), syne1[e]) for e in edges1 if nodes1[i] in e]))

  for i in range(len(node_mapping[1])):
    if node_mapping[1][i] != i:
       if node_mapping[1][i] != None:
        try:
          if synn2[nodes2[i]] != synn1[nodes1[node_mapping[1][i]]]:
            node_edits2+=1
            node_edits2_.append((synn2[nodes2[i]][0], synn1[nodes1[node_mapping[1][i]]][0], [((synn2[e[0]][0], synn2[e[1]][0]), syne2[e]) for e in edges2 if nodes2[i] in e]))
        except:
            node_edits2+=1
            # print(node_mapping[1], len(node_mapping[1]), len(nodes1), len(nodes2))
            node_edits2_.append((None, synn1[nodes1[node_mapping[1][i]]][0], [((synn2[e[0]][0], synn2[e[1]][0]), syne2[e]) for e in edges2 if nodes2[i] in e]))
       elif node_mapping[1][i] == None:
        node_edits2+=1
        node_edits2_.append((synn2[nodes2[i]][0], None, [((synn2[e[0]][0], synn2[e[1]][0]), syne2[e]) for e in edges2 if nodes2[i] in e]))

  for i in range(len(edge_mapping[0])):
    if edge_mapping[0][i] != i:
      if edge_mapping[0][i] != None:
        try:
          if syne1[edges1[i]] != syne2[edges2[edge_mapping[0][i]]]:
            edge_edits1+=1
            edge_edits1_.append((syne1[edges1[i]], syne2[edges2[edge_mapping[0][i]]], (synn1[edges1[i][0]][0], synn1[edges1[i][1]][0])))
        except:
          edge_edits1+=1
          edge_edits1_.append((None, syne2[edges2[edge_mapping[0][i]]][0], (synn1[edges1[i][0]][0], synn1[edges1[i][1]][0])))
      elif edge_mapping[0][i] == None:
        edge_edits1+=1
        edge_edits1_.append((syne1[edges1[i]], None, (synn1[edges1[i][0]][0], synn1[edges1[i][1]][0])))

  for i in range(len(edge_mapping[1])):
    if edge_mapping[1][i] != i:
      if edge_mapping[1][i] != None:
        try:
          if syne2[edges2[i]] != syne1[edges1[edge_mapping[1][i]]]:
            edge_edits2+=1
            edge_edits2_.append((syne2[edges2[i]], syne1[edges1[edge_mapping[1][i]]], (synn2[edges2[i][0]][0], synn2[edges2[i][1]][0])))
        except:
          edge_edits2+=1
          edge_edits2_.append((None, syne1[edges1[edge_mapping[1][i]]], (synn2[edges2[i][0]][0], synn2[edges2[i][1]][0])))
      elif edge_mapping[1][i] == None:
        edge_edits2+=1
        edge_edits2_.append((syne2[edges2[i]], None, (synn2[edges2[i][0]][0], synn2[edges2[i][1]][0])))

  return node_edits1, node_edits2, edge_edits1, edge_edits2 , {'node1': node_edits1_, 'node2': node_edits2_, 'edge1': edge_edits1_, 'edge2': edge_edits2_}
```

```python
 def find_edits(pairs_sgg, geds, graphs_sgg_ids, graphs_sgg):
  edits = {}
  edits_path = {}

  for pair in pairs_sgg:
    if pair in geds.keys():
      _, nodem, edgem = geds[pair]
      n, _, e, _ , d = count_edits(pair, nodem, edgem, graphs_sgg_ids, graphs_sgg)
    elif (pair[1], pair[0]) in geds.keys():
      # print('yes')
      pair_ = (pair[1], pair[0])
      _, nodem, edgem = geds[pair_]
      _, n, _, e, d = count_edits(pair_, nodem, edgem, graphs_sgg_ids, graphs_sgg)
      d_ = {'node1': d['node2'], 'node2': d['node1'], 'edge1': d['edge2'], 'edge2':d['edge1']}
      d = d_
    else:
      print(pair)
      continue

    edits[pair] = (n,e)
    edits_path[pair] = d

  return edits, edits_path
```

```python
from collections import Counter

def get_globals(paths, smarty=False):
  node_edits = []
  edge_edits = []
  edge_single = []
  triple_edits = []
  for k, v in paths.items():
    nes = [(n[0], n[1]) for n in v['node1']] + [(n[1], n[0]) for n in v['node2'] if None in n]
    ees = v['edge1'] + [(n[1], n[0], n[2]) for n in v['edge2'] if None in n]

    if smarty:
      candidate_swaps = [ne for ne in nes if None not in ne]
      right =[ne[1] for ne in candidate_swaps]
      left =[ne[0] for ne in candidate_swaps]

      new_swaps = []
      for idx, l in enumerate(left):
        found = False
        for g in groups:
          if l in g:
            g_ = g - {l}
            for other in g_:
              if other in right:
                new_swaps.append((l, other))
                found=True
                break
            if not found:
              new_swaps.append((left[idx], right[idx]))
        if not found:
          new_swaps.append((left[idx], right[idx]))
      cnt = 0
      for idx, ne in enumerate(nes):
        if None not in ne:
            nes[idx] = new_swaps[cnt]
            cnt+=1



    node_edits.append(nes)
    edge_edits.append(ees)

    edge_single.append([(e[0], e[1]) for e in v['edge1'] + [(n[1], n[0], n[2]) for n in v['edge2'] if None in n]])


    for edge in v['edge1']:
      nodes = edge[2]
      n1 = nodes[0]
      n2 = nodes[1]
      for node_pair in v['node1']:
        if node_pair[0] == nodes[0]:
          n1 = node_pair[1]
        elif node_pair[0] == nodes[1]:
          n2 = node_pair[1]
      if None in edge:
        triple = ((nodes[0], edge[0], nodes[1]), None)
      else:
        triple = ((nodes[0], edge[0], nodes[1]), (n1, edge[1], n2))
      triple_edits.append(triple)

    for edge in v['edge2']:
      nodes = edge[2]
      n1 = nodes[0]
      n2 = nodes[1]
      if edge[1] == None:
        for node_pair in v['node2']:
          if node_pair[0] == nodes[0]:
            n1 = node_pair[1]
          elif node_pair[0] == nodes[1]:
            n2 = node_pair[1]
        if None in edge:
          triple = (None, (nodes[0], edge[0], nodes[1]))
        else:
          continue
        triple_edits.append(triple)


  c1 = Counter([item for sublist in node_edits for item in sublist])
  c2 = Counter([item for sublist in edge_edits for item in sublist])
  c3 = Counter([item for sublist in edge_single for item in sublist])
  c4 = Counter(triple_edits)

  d1 = dict(sorted(c1.items(), key=lambda item: item[1], reverse=True))
  d2 = dict(sorted(c2.items(), key=lambda item: item[1], reverse=True))
  d3 = dict(sorted(c3.items(), key=lambda item: item[1], reverse=True))
  d4 = dict(sorted(c4.items(), key=lambda item: item[1], reverse=True))
  print(d1)
  print(d2)
  print(d3)
  print(d4)

  return d1, d2, d3, d4
```

```python
def add_dicts(dict1, dict2):
  dd = {k : dict2[k] + v if k in dict2 else v  for k,v in dict1.items()}
  return dict(sorted(dd.items(), key=lambda item: item[1], reverse=True))
```

```python
import matplotlib.pyplot as plt
import math

def plot_triple_counts(data_dict, fig_size, title, length, what='Triples'):

  labels = {}

  for c, v in data_dict.items():
    if c[0] == None:
      if str(c[1]) in labels:
         color, value = labels[str(c[1])]
         new_value = abs(value - v)
         new_color = color if value - v > 0 else 'green'
         labels[str(c[1])] = (new_color, new_value)
      else:
        labels[str(c[1])] = ('green', v)
    elif c[1] == None:
      if str(c[0]) in labels:
         color, value = labels[str(c[0])]
         new_value = abs(value - v)
         new_color = color if value - v > 0 else 'red'
         labels[str(c[0])] = (new_color, new_value)
      else:
        labels[str(c[0])] = ('red', v)
    else:
      if str((c[1], c[0])) in labels:
         color, value = labels[str((c[1], c[0]))]
         new_value = abs(value - v)
         new_label = str((c[1], c[0])) if value - v > 0 else str(c)
         labels[new_label] = ('blue', new_value)
      else:
        labels[str(c)] = ('blue', v)


  labels = dict(sorted(labels.items(), key=lambda item: (-item[1][1], item[0])))

  categories = list(labels.keys())[:length]
  counts = [v[1] / max(v[1] for v in list(labels.values())[:length]) for v in labels.values()][:length]
  colors = [v[0] for v in labels.values()][:length]

  # print(labels)

  # Create a bar plot
  plt.figure(figsize=fig_size)
  plt.bar(categories, counts, color=colors)

  color_mapping = {'add': 'green', 'delete': 'red', 'swap': 'blue'}
  inverse_color_mapping = {v:k for k, v in color_mapping.items()}
  legend_labels = [plt.Rectangle((0,0),1,1, color=color_mapping[category]) for category in [inverse_color_mapping[c] for c in set(colors)]]
  plt.legend(legend_labels, [inverse_color_mapping[c] for c in set(colors)], fontsize=15)
  # Label the axes and add a title
  plt.xlabel(what, fontsize=20)
  plt.ylabel('Normalized Counts', fontsize=20)
  plt.title(title)

  plt.xticks(rotation=65, fontsize=20, ha='right' )
  # plt.xticks(rotation='vertical',fontsize=13)

  print({i: v[1]/ max(v[1] for v in list(labels.values())[:length]) for i,v in labels.items()})
  # Show the plot
  plt.show()
```
